## Python Descriptive Statistics - Continuous Integration 

##### The dataset we're working with is about top hits on spotify. It's stored in this directory as    ```spotify.csv```

````python
import lib
````

````python
#reading dataset
spotify = lib.reader("spotify.csv")
#to check how it looks - checking first 10 rows
spotify.head(10)
````

````python
#describing the dataset - to get a sense of the values 
spotify.describe()
````

### Descriptive statistics of the duration of the top hits on spotify

````python
#descriptive statistics on the duration of songs
column = "duration_ms"
mean = lib.mean (column, spotify)
median = lib.median (column, spotify)
mode = lib.mode (column, spotify)
std = lib.std (column, spotify)

print ("Mean duration of songs:", mean)
print ("Median duration of songs:", median)
print ("Mode duration of songs:", mode)
print ("std duration of songs:", std)
````

### Who were the top 10 artists with most top hits on spotify between 2012 and 2022 (Visualization)

````python
# Plotting 
x = lib.viz("artist_name", spotify) # creates visualization and saves it in the output folder
x
````
