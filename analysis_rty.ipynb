```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load and prepare data
df = pd.read_csv("rty.csv")
df["DateTime"] = pd.to_datetime(df["Date"] + " " + df["TimeStart"])
df["Date"] = pd.to_datetime(df["Date"])


# Function to calculate returns for a specific time range
def calculate_returns(df, start_time, end_time):
    mask = (df["DateTime"].dt.time >= start_time) & (df["DateTime"].dt.time < end_time)
    return df[mask].groupby("Date")["LogReturn"].sum()


# Calculate returns for two time slots
returns_1 = calculate_returns(
    df, pd.Timestamp("09:00").time(), pd.Timestamp("09:30").time()
)
returns_2 = calculate_returns(
    df, pd.Timestamp("09:30").time(), pd.Timestamp("10:00").time()
)

# Create DataFrame for plotting
plot_df = pd.DataFrame({"Returns_1": returns_1, "Returns_2": returns_2})
plot_df["DaysAgo"] = (plot_df.index.max() - plot_df.index).days

# Create interactive scatter plot
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=plot_df["Returns_1"],
        y=plot_df["Returns_2"],
        mode="markers",
        marker=dict(
            size=10,
            color=plot_df["DaysAgo"],
            colorscale="Viridis",
            colorbar=dict(title="Days Ago"),
        ),
        text=plot_df.index.strftime("%Y-%m-%d"),
        hovertemplate="Date: %{text}<br>Returns 9:00-9:30: %{x:.4f}<br>Returns 9:30-10:00: %{y:.4f}<extra></extra>",
    )
)

fig.update_layout(
    title="RTY Futures Returns: 9:00-9:30 vs 9:30-10:00",
    xaxis_title="Returns 9:00-9:30",
    yaxis_title="Returns 9:30-10:00",
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import numpy as np
from scipy import stats

# Load and prepare data (assuming this part works as before)
df = pd.read_csv("rty.csv")


def parse_datetime(date, time_start):
    date = pd.to_datetime(date)
    time_parts = time_start.split()
    if len(time_parts) == 3:  # Format: "0 days HH:MM:SS.fraction"
        time = pd.to_timedelta(time_parts[2])
    else:  # Format: "HH:MM:SS.fraction"
        time = pd.to_timedelta(time_parts[0])
    return date + time


# Apply the parsing function
df["DateTime"] = df.apply(
    lambda row: parse_datetime(row["Date"], row["TimeStart"]), axis=1
)
df["Date"] = pd.to_datetime(df["Date"])


def calculate_returns(df, start_time, end_time):
    mask = (df["DateTime"].dt.time >= start_time) & (df["DateTime"].dt.time < end_time)
    return df[mask].groupby("Date")["LogReturn"].sum()


app = dash.Dash(__name__)

app.layout = html.Div(
    [
        html.H1("RTY Futures Returns Analysis"),
        html.Div(
            [
                html.Div(
                    [
                        html.Label("Time Slot 1:"),
                        dcc.Input(
                            id="time1-start",
                            type="text",
                            value="09:00",
                            style={"marginRight": "10px"},
                        ),
                        dcc.Input(id="time1-end", type="text", value="09:30"),
                    ],
                    style={"marginBottom": "10px"},
                ),
                html.Div(
                    [
                        html.Label("Time Slot 2:"),
                        dcc.Input(
                            id="time2-start",
                            type="text",
                            value="09:30",
                            style={"marginRight": "10px"},
                        ),
                        dcc.Input(id="time2-end", type="text", value="10:00"),
                    ],
                    style={"marginBottom": "10px"},
                ),
                html.Button("Update", id="update-button", n_clicks=0),
            ],
            style={"marginBottom": "20px"},
        ),
        dcc.Graph(id="returns-plot"),
        html.Div(
            [
                html.Div(
                    [dcc.Graph(id="distribution-plot")],
                    style={"width": "50%", "display": "inline-block"},
                ),
                html.Div(
                    [dcc.Graph(id="heatmap-plot")],
                    style={"width": "50%", "display": "inline-block"},
                ),
            ]
        ),
        html.Div(id="stats-output"),
    ]
)


@app.callback(
    [
        Output("returns-plot", "figure"),
        Output("distribution-plot", "figure"),
        Output("heatmap-plot", "figure"),
        Output("stats-output", "children"),
    ],
    [Input("update-button", "n_clicks")],
    [
        State("time1-start", "value"),
        State("time1-end", "value"),
        State("time2-start", "value"),
        State("time2-end", "value"),
    ],
)
def update_plots(n_clicks, time1_start, time1_end, time2_start, time2_end):
    returns_1 = calculate_returns(
        df, pd.to_datetime(time1_start).time(), pd.to_datetime(time1_end).time()
    )
    returns_2 = calculate_returns(
        df, pd.to_datetime(time2_start).time(), pd.to_datetime(time2_end).time()
    )

    plot_df = pd.DataFrame({"Returns_1": returns_1, "Returns_2": returns_2})
    plot_df["DaysAgo"] = (plot_df.index.max() - plot_df.index).days

    # Scatter plot
    fig1 = go.Figure()
    fig1.add_trace(
        go.Scatter(
            x=plot_df["Returns_1"],
            y=plot_df["Returns_2"],
            mode="markers",
            marker=dict(
                size=10,
                color=plot_df["DaysAgo"],
                colorscale="Viridis",
                colorbar=dict(title="Days Ago"),
            ),
            text=plot_df.index.strftime("%Y-%m-%d"),
            hovertemplate="Date: %{text}<br>Returns "
            + time1_start
            + "-"
            + time1_end
            + ": %{x:.4f}<br>Returns "
            + time2_start
            + "-"
            + time2_end
            + ": %{y:.4f}<extra></extra>",
        )
    )

    # Add regression line
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        plot_df["Returns_1"], plot_df["Returns_2"]
    )
    line_x = np.array([plot_df["Returns_1"].min(), plot_df["Returns_1"].max()])
    line_y = slope * line_x + intercept

    fig1.add_trace(go.Scatter(x=line_x, y=line_y, mode="lines", name="Regression Line"))

    fig1.update_layout(
        title=f"RTY Futures Returns: {time1_start}-{time1_end} vs {time2_start}-{time2_end}",
        xaxis_title=f"Returns {time1_start}-{time1_end}",
        yaxis_title=f"Returns {time2_start}-{time2_end}",
    )

    # Distribution plot
    fig2 = go.Figure()
    fig2.add_trace(
        go.Histogram(x=plot_df["Returns_1"], name=f"{time1_start}-{time1_end}")
    )
    fig2.add_trace(
        go.Histogram(x=plot_df["Returns_2"], name=f"{time2_start}-{time2_end}")
    )
    fig2.update_layout(title="Distribution of Returns", barmode="overlay")

    # Heatmap of day-of-week vs hour
    plot_df["DayOfWeek"] = plot_df.index.dayofweek
    plot_df["Hour"] = plot_df.index.hour
    heatmap_data = plot_df.pivot_table(
        values="Returns_1", index="DayOfWeek", columns="Hour", aggfunc="mean"
    )
    fig3 = px.imshow(heatmap_data, title="Average Returns by Day and Hour")

    # Calculate statistics
    correlation = plot_df["Returns_1"].corr(plot_df["Returns_2"])
    sharpe_ratio_1 = (
        np.sqrt(252) * plot_df["Returns_1"].mean() / plot_df["Returns_1"].std()
    )
    sharpe_ratio_2 = (
        np.sqrt(252) * plot_df["Returns_2"].mean() / plot_df["Returns_2"].std()
    )

    stats_output = html.Div(
        [
            html.H3("Statistics:"),
            html.P(f"Correlation: {correlation:.4f}"),
            html.P(f"R-squared: {r_value**2:.4f}"),
            html.P(f"Regression Slope: {slope:.4f}"),
            html.P(f"Regression Intercept: {intercept:.4f}"),
            html.P(f"Sharpe Ratio ({time1_start}-{time1_end}): {sharpe_ratio_1:.4f}"),
            html.P(f"Sharpe Ratio ({time2_start}-{time2_end}): {sharpe_ratio_2:.4f}"),
        ]
    )

    return fig1, fig2, fig3, stats_output


if __name__ == "__main__":
    app.run_server(debug=True)
```

```python
import streamlit as st
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
from scipy import stats


# Load and prepare data
@st.cache_data
def load_data():
    df = pd.read_csv("rty.csv")

    def parse_datetime(date, time_start):
        date = pd.to_datetime(date)
        time_parts = time_start.split()
        if len(time_parts) == 3:  # Format: "0 days HH:MM:SS.fraction"
            time = pd.to_timedelta(time_parts[2])
        else:  # Format: "HH:MM:SS.fraction"
            time = pd.to_timedelta(time_parts[0])
        return date + time

    df["DateTime"] = df.apply(
        lambda row: parse_datetime(row["Date"], row["TimeStart"]), axis=1
    )
    df["Date"] = pd.to_datetime(df["Date"])
    return df


df = load_data()


def calculate_returns(df, start_time, end_time):
    mask = (df["DateTime"].dt.time >= start_time) & (df["DateTime"].dt.time < end_time)
    return df[mask].groupby("Date")["LogReturn"].sum()


# Streamlit app
st.title("RTY Futures Returns Analysis")

# Time slot inputs
col1, col2 = st.columns(2)
with col1:
    time1_start = st.time_input(
        "Time Slot 1 Start", value=pd.to_datetime("09:00").time()
    )
    time1_end = st.time_input("Time Slot 1 End", value=pd.to_datetime("09:30").time())
with col2:
    time2_start = st.time_input(
        "Time Slot 2 Start", value=pd.to_datetime("09:30").time()
    )
    time2_end = st.time_input("Time Slot 2 End", value=pd.to_datetime("10:00").time())

if st.button("Update Analysis"):
    returns_1 = calculate_returns(df, time1_start, time1_end)
    returns_2 = calculate_returns(df, time2_start, time2_end)

    plot_df = pd.DataFrame({"Returns_1": returns_1, "Returns_2": returns_2})
    plot_df["DaysAgo"] = (plot_df.index.max() - plot_df.index).days

    # Scatter plot
    fig1 = go.Figure()
    fig1.add_trace(
        go.Scatter(
            x=plot_df["Returns_1"],
            y=plot_df["Returns_2"],
            mode="markers",
            marker=dict(
                size=10,
                color=plot_df["DaysAgo"],
                colorscale="Viridis",
                colorbar=dict(title="Days Ago"),
            ),
            text=plot_df.index.strftime("%Y-%m-%d"),
            hovertemplate="Date: %{text}<br>Returns "
            + time1_start.strftime("%H:%M")
            + "-"
            + time1_end.strftime("%H:%M")
            + ": %{x:.4f}<br>Returns "
            + time2_start.strftime("%H:%M")
            + "-"
            + time2_end.strftime("%H:%M")
            + ": %{y:.4f}<extra></extra>",
        )
    )

    # Add regression line
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        plot_df["Returns_1"], plot_df["Returns_2"]
    )
    line_x = np.array([plot_df["Returns_1"].min(), plot_df["Returns_1"].max()])
    line_y = slope * line_x + intercept

    fig1.add_trace(go.Scatter(x=line_x, y=line_y, mode="lines", name="Regression Line"))

    fig1.update_layout(
        title=f'RTY Futures Returns: {time1_start.strftime("%H:%M")}-{time1_end.strftime("%H:%M")} vs {time2_start.strftime("%H:%M")}-{time2_end.strftime("%H:%M")}',
        xaxis_title=f'Returns {time1_start.strftime("%H:%M")}-{time1_end.strftime("%H:%M")}',
        yaxis_title=f'Returns {time2_start.strftime("%H:%M")}-{time2_end.strftime("%H:%M")}',
    )

    st.plotly_chart(fig1)

    # Distribution plot
    fig2 = go.Figure()
    fig2.add_trace(
        go.Histogram(
            x=plot_df["Returns_1"],
            name=f'{time1_start.strftime("%H:%M")}-{time1_end.strftime("%H:%M")}',
        )
    )
    fig2.add_trace(
        go.Histogram(
            x=plot_df["Returns_2"],
            name=f'{time2_start.strftime("%H:%M")}-{time2_end.strftime("%H:%M")}',
        )
    )
    fig2.update_layout(title="Distribution of Returns", barmode="overlay")

    st.plotly_chart(fig2)

    # Heatmap of day-of-week vs hour
    plot_df["DayOfWeek"] = plot_df.index.dayofweek
    plot_df["Hour"] = plot_df.index.hour
    heatmap_data = plot_df.pivot_table(
        values="Returns_1", index="DayOfWeek", columns="Hour", aggfunc="mean"
    )
    fig3 = px.imshow(heatmap_data, title="Average Returns by Day and Hour")

    st.plotly_chart(fig3)

    # Calculate and display statistics
    correlation = plot_df["Returns_1"].corr(plot_df["Returns_2"])
    sharpe_ratio_1 = (
        np.sqrt(252) * plot_df["Returns_1"].mean() / plot_df["Returns_1"].std()
    )
    sharpe_ratio_2 = (
        np.sqrt(252) * plot_df["Returns_2"].mean() / plot_df["Returns_2"].std()
    )

    st.subheader("Statistics:")
    st.write(f"Correlation: {correlation:.4f}")
    st.write(f"R-squared: {r_value**2:.4f}")
    st.write(f"Regression Slope: {slope:.4f}")
    st.write(f"Regression Intercept: {intercept:.4f}")
    st.write(
        f"Sharpe Ratio ({time1_start.strftime('%H:%M')}-{time1_end.strftime('%H:%M')}): {sharpe_ratio_1:.4f}"
    )
    st.write(
        f"Sharpe Ratio ({time2_start.strftime('%H:%M')}-{time2_end.strftime('%H:%M')}): {sharpe_ratio_2:.4f}"
    )
```
